```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
directory = '../data/'
filename_2020_part1 = 'HIST_PAINEL_COVIDBR_2020_Parte1_04mai2022.csv'
filename_2020_part2 = 'HIST_PAINEL_COVIDBR_2020_Parte2_04mai2022.csv'
filename_2021_part1 = 'HIST_PAINEL_COVIDBR_2021_Parte1_04mai2022.csv'
filename_2021_part2 = 'HIST_PAINEL_COVIDBR_2021_Parte2_04mai2022.csv'
filename_2022_part1 = 'HIST_PAINEL_COVIDBR_2022_Parte1_04mai2022.csv'



df1 = pd.read_csv(directory+filename_2020_part1, sep=';')
df2 = pd.read_csv(directory+filename_2020_part2, sep=';')
df3 = pd.read_csv(directory+filename_2021_part1, sep=';')
df4 = pd.read_csv(directory+filename_2021_part2, sep=';')
df5 = pd.read_csv(directory+filename_2022_part1, sep=';')
```

```python
df = pd.concat([df1, df2, df3, df4, df5])
df
```

```python
# df = df.sort_values(by='data')
# df
```

```python
df = df.reset_index().drop(['index'], axis=1)
df
```
