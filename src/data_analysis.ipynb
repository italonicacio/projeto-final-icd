## KMeans

```python
from sklearn import preprocessing
from sklearn.cluster import KMeans

df.info()
```

```python
df.sort_values(by='achievements',ascending=False).head(10)
```

```python
from collections import defaultdict
d = defaultdict(preprocessing.LabelEncoder)

fit = df.apply(lambda x: d[x.name].fit_transform(x))

fit.apply(lambda x: d[x.name].inverse_transform(x))

df = df.apply(lambda x: d[x.name].transform(x))

df['price'] = d['price'].inverse_transform(df['price'])

```

```python
clf = KMeans(n_clusters=10)

col = ['name', 'publisher', 'price', 'categories', 'platforms', 'achievements']

clf.fit(df[col])
# clf.fit(df)

```

## Regressão Linear

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
y = df['price']
# new_df = df.loc[:, ~df.columns.isin(['price'])]
new_df = df[['publisher', 'genres']]

x_train, x_test, y_train, y_test = train_test_split(new_df, y, test_size=0.20)
```

```python
x_train
```

```python
lr = LinearRegression(n_jobs=-1)

lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)
```

```python
lr.intercept_
```

```python
size = 500
columns_name = 'publisher'
columns_id = 0
x = np.linspace(-1, 500, 1000)

plt.scatter(x_test[columns_name][:size], x_test['genres'][:size], c=y_test[:size])
plt.plot(x_test[columns_name][:size], lr.coef_[columns_id]*x_test[columns_name][:size] + lr.intercept_, c='blue')
```
